```python
import json, collections

path = r"C:\Users\Administrator\Documents\URBENCH\data\samples\urbench_v0_1.jsonl"

with open(path, "r", encoding="utf-8") as f:
    rows = [json.loads(line) for line in f]

print("Total items:", len(rows))

required = {"id","category","question_ur","answer_ur","answer_norm"}
missing = [r.get("id","?") for r in rows if not required.issubset(r)]
print("Missing-field items:", missing)

q_counts = collections.Counter(r["question_ur"].strip() for r in rows)
dupes = [q for q,c in q_counts.items() if c>1]
print("Duplicate questions:", len(dupes))

by_cat = collections.Counter(r["category"] for r in rows)
print("Counts by category:", dict(by_cat))

for i, r in enumerate(rows[:5], 1):
    print(f"{i}. {r['question_ur']} => {r['answer_norm']}")
```
