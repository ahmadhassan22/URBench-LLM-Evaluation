```python
import json
from pathlib import Path

# 1) Paths
en_path = Path("../data/source_en/commonsenseqa_en_sample_v0.jsonl")
ur_path = Path("../data/translated_ur/commonsenseqa_ur_v0.jsonl")

ur_path.parent.mkdir(parents=True, exist_ok=True)

print("Reading from:", en_path.resolve())
print("Writing to  :", ur_path.resolve())

# 2) Helper: from answer label ('A','B',...) to English text
def label_to_text(rec):
    labels = rec["choice_labels"]
    texts = rec["choices_en"]
    mapping = {lab: txt for lab, txt in zip(labels, texts)}
    return mapping[rec["answer_label"]]

# 3) Build Urdu template
count = 0
with en_path.open("r", encoding="utf-8") as f_in, ur_path.open("w", encoding="utf-8") as f_out:
    for line in f_in:
        rec = json.loads(line)
        answer_en = label_to_text(rec)

        out = {
            "id": rec["id"],
            "source_id": rec["source_id"],
            "category": rec["category"],

            "question_en": rec["question_en"],
            "question_ur": "",          # 👉 YOU will fill

            "choices_en": rec["choices_en"],
            "choices_ur": [""] * len(rec["choices_en"]),  # 👉 YOU will fill

            "choice_labels": rec["choice_labels"],  # ['A','B','C','D','E']

            "answer_label": rec["answer_label"],    # e.g. 'A'
            "answer_en": answer_en,
            "answer_ur": ""          # 👉 YOU will fill
        }

        f_out.write(json.dumps(out, ensure_ascii=False) + "\n")
        count += 1

print("Template rows written:", count)
```
