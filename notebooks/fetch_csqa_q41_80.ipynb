```python
from datasets import load_dataset
import json
import os

# Load dataset
ds = load_dataset("tau/commonsense_qa")
train = ds["train"]

# Check one example structure
print(train[0])

# Fetch next 40 items (indices 40 to 79)
start = 40
end = 80

print("Total train size:", len(train))
print(f"Fetching items {start} to {end-1}")

# Output path
out_dir = "data/source_en"
os.makedirs(out_dir, exist_ok=True)

out_path = os.path.join(out_dir, "commonsenseqa_en_q41_80.jsonl")

# Write JSONL
count = 0
with open(out_path, "w", encoding="utf-8") as f_out:
    for rec in train[start:end]:
        out = {
            "id": f"CSQA_{start+count+1:04d}",
            "question_en": rec["question"],
            "choice_labels": rec["choices"]["label"],
            "choice_text_en": rec["choices"]["text"],
            "answer_label": rec["answerKey"]
        }
        f_out.write(json.dumps(out, ensure_ascii=False) + "\n")
        count += 1

print("Written questions:", count)
print("Saved to:", out_path)
```
