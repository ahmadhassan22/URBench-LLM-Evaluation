```python
from datasets import load_dataset

# 1. Load official CommonsenseQA dataset (English)
ds = load_dataset("tau/commonsense_qa")

# 2. Select training split
train = ds["train"]

# 3. Display dataset size and the first item
print("Train size:", len(train))
train[0]
```

```python
import json
from pathlib import Path

# Use the same `train` we loaded above
print("Train size:", len(train))

# How many items you want in the first English batch
N = 40

# Make sure folder exists: ../data/source_en/
out_path = Path("../data/source_en/commonsenseqa_en_sample_v0.jsonl")
out_path.parent.mkdir(parents=True, exist_ok=True)

with out_path.open("w", encoding="utf-8") as f:
    for i in range(N):
        ex = train[i]
        # Build a simple, clean structure
        record = {
            "id": f"CSQA-{i+1:03d}",
            "source_id": ex["id"],
            "category": "commonsense",
            "question_en": ex["question"],
            "choices_en": ex["choices"]["text"],
            "choice_labels": ex["choices"]["label"],
            "answer_label": ex["answerKey"],   # e.g. 'A'
        }
        f.write(json.dumps(record, ensure_ascii=False) + "\n")

print("Saved", N, "items to:", out_path.resolve())
```
